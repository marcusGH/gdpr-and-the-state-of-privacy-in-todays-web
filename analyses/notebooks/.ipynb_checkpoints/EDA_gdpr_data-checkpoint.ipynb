```r
df <- read.csv(here("data", "gdpr_compliancy_data_full.csv"))
df$tld <- toupper(df$tld)
head(df)
```

```r
eu_tlds <- eu_countries %>%
    select(code) %>%
    pull()

tibble(df) %>%
    filter(tld %in% eu_tlds) %>%
    mutate(proportion_violation = num_google_ads_violations / num_requests) %>%
    select(tld, proportion_violation) %>%
    rename(geo = tld) %>%
    inner_join(eu_shape, by = "geo") %>%
    st_as_sf() %>%
    # Plotting starts
    ggplot(aes(fill = proportion_violation)) +
    geom_sf(
        size = 1,
        color = "#fdbf11"
    ) +
    scale_fill_continuous(
        # name = "Proportion of traffic that violates GDPR",
        name = "",
        labels = label_percent(),
        breaks = pretty_breaks(6),
        guide = guide_colorbar(
            direction = "horizontal",
            title.position = "top",
            label.position = "bottom",
            barwidth = unit(12, "cm"),
            barheight = unit(.5, "cm"),
            ticks = TRUE,
            ticks.linewidth = 1,
            nbin=500,
        ),
        limits = c(0, NA)
    ) +
    theme_void() +
    theme(legend.position = "bottom") +
    scale_x_continuous(limits = c(-10, 35)) +
    scale_y_continuous(limits = c(35, 65)) +
    labs(
        title = "Proportion of traffic that violates GDPR in EU",
        subtitle = "Measures in % of requests sent in the first two weeks of January",
        caption = "Data: httparchive"
    )
```

```r
tibble(df) %>%
    filter(tld %in% eu_tlds) %>%
    mutate(proportion_violation = num_google_ads_violations / num_requests) %>%
    rename(code = tld) %>%
    # to get the country names
    left_join(eu_countries, by = "code") %>%
    select(name, proportion_violation) %>%
    arrange(proportion_violation) %>%
    # ensure the countries will appear in sorted order
    mutate(name = factor(name, levels = .$name)) %>%
    ggplot(aes(proportion_violation, name)) +
        geom_segment(aes(x = 0, xend = proportion_violation, y = name, yend = name)) +
        geom_point() +
        scale_x_continuous(expand = expansion(mult = c(0, 0)), limits = c(0, 0.063)) +
        labs(x = NULL,
             y = "% of traffic that violates GDPR")
```

# Waffle plot

```r
head(df)
```

```r
# The different categories can be seperates into 4 disjoint sets:
#  +-----------------------+
#  |      /-----+-----\    |
#  |  A  (  B / D \ C )    |
#  |     \----+-+-+---/    |
#  +-----------------------+
#
#  where:
#   D = websites using google adsense and has a cookie compliance banner
#   B = websites using a cookie compliance banner, but without google ads
#   C = websites using google ads, but without cookie compliance banner (GDPR violation)
#   A = websites doing none of the above
tibble(df) %>%
    filter(tld %in% eu_tlds) %>%
    rename(geo = tld) %>%
    mutate(C = num_unique_pages_with_google_ads - num_unique_pages_legal_google_ads) %>%
    mutate(B = num_unique_pages_with_cookie_compliance - num_unique_pages_legal_google_ads) %>%
    mutate(D = num_unique_pages_legal_google_ads) %>%
    mutate(A = num_unique_pages - B - C - D) %>%
    summarise(A = sum(A), B = sum(B), D = sum(D), C = sum(C)) %>%
    gather() %>%
    # plot the data
    ggplot(aes(fill = key, values = value)) +
        geom_waffle(n_rows = 10, make_proportional = TRUE)
```

```r
tibble(df) %>%
    filter(tld %in% c("COM", "NET", "ORG")) %>%
    rename(geo = tld) %>%
    mutate(C = num_unique_pages_with_google_ads - num_unique_pages_legal_google_ads) %>%
    mutate(B = num_unique_pages_with_cookie_compliance - num_unique_pages_legal_google_ads) %>%
    mutate(D = num_unique_pages_legal_google_ads) %>%
    mutate(A = num_unique_pages - B - C - D) %>%
    summarise(A = sum(A), B = sum(B),  D = sum(D), C = sum(C)) %>%
    gather() %>%
    # plot the data
    ggplot(aes(fill = key, values = value)) +
        geom_waffle(n_rows = 10, make_proportional = TRUE)
```

```r
# The different categories can be seperates into 4 disjoint sets:
#  +-----------------------+
#  |      /-----+-----\    |
#  |  A  (  B / C \ D )    |
#  |     \----+-+-+---/    |
#  +-----------------------+
#
#  where:
#   C = websites using google adsense and has a cookie compliance banner
#   B = websites using a cookie compliance banner, but without google ads
#   D = websites using google ads, but without cookie compliance banner (GDPR violation)
#   A = websites doing none of the above
#  and we currently have:
#   num_unique_pages_legal_google_ads        = C
#   num_unique_pages_with_google_ads         = C + D
#   num_unique_pages_with_cookie_compliance  = B + C
#   num_unique_pages                         = A + B + C + D
# divide all the numbers by this value
quantity = 10000
tibble(df) %>%
    # filter(tld %in% eu_tlds) %>%
    mutate(D = num_unique_pages_with_google_ads - num_unique_pages_legal_google_ads) %>%
    mutate(B = num_unique_pages_with_cookie_compliance - num_unique_pages_legal_google_ads) %>%
    mutate(C = num_unique_pages_legal_google_ads) %>%
    mutate(A = num_unique_pages - B - C - D) %>%
    select(tld, A, B, C, D) %>%
    mutate(location = ifelse(tld %in% eu_tlds, "Europe", ifelse(tld %in% c("ORG", "NET", "COM"), "International", "Other"))) %>%
    group_by(location) %>%
    summarise(D = sum(D) / quantity,
              C = sum(C) / quantity,
              B = sum(B) / quantity,
              A = sum(A) / quantity) %>%
    pivot_longer(cols = !location) -> waffle_data
waffle_data
```

```r
# create the plot
waffle_data %>%
    ggplot(aes(fill=name, values=value)) +
    geom_waffle(size = .25, n_rows = 10, flip = TRUE) +
    facet_wrap(~location, nrow = 1, strip.position = "bottom") +
    scale_x_discrete() +
    scale_y_continuous(labels = function(x) x * quantity,
                       expand = c(0, 0)) +
    coord_equal() +
    labs(
        title = "test",
        subtitle = "sub",
        x = "Top-level domain category",
        y = "Number of unique websites"
    ) +
    theme_minimal(base_family = "Lato") +
    theme(panel.grid = element_blank(), axis.ticks.y = element_line()) +
    guides(fill = guide_legend(reverse = TRUE))
    
# TODO: make this for the total number of websites instead
# Then make proportion bar plot comparing the three categories of websites to highlight proportions
# also, with cateogires: EU-country, other-country TLD, international, other
```
